### Defining a few functions to help us write lesser code

```python
def lexical_diversity(text):
  return len(text) / len(set(text))
```

```python
def percentage(count,total):
  return 100 * count / total
```

```python
lexical_diversity(text4)
```

```python
lexical_diversity(text5)
```

```python
percentage(4,5)
```

```python
percentage(text4.count('the'),len(text4))

# as we see, this matches the output we had gotten earlier too so our functions are working good
```

# Frequency Distributions

```python
# frequency distribution , and it tells us the
# frequency of each vocabulary item in the text. (In general, it could count any kind of
# observable event.) It is a distribution since it tells us how the total number of word
# tokens in the text are distributed across the vocabulary items.

fdist1 = FreqDist(text1)
```

```python
fdist1
```

```python
fdist1['whale']
```
